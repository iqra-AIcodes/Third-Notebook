```python
# Task 4: Simple Function
# A function is a block of code we can reuse by calling its name

# Step 1: Define the function
def welcome(name):
    # This function prints a greeting message
    print("Hello", name + ", welcome to Python!")

# Step 2: Call the function multiple times
welcome("Iqra Eiman")
welcome("Ali")
welcome("Sara")
```

```python
# Task 4: Simple Function with user input

# Step 1: Define the function
def welcome(name):
    # This function prints a greeting message
    print("Hello", name + ", welcome to Python!")

# Step 2: Ask user for names and call function
for i in range(3):  # we will take 3 names
    user_name = input("Enter a name: ")
    welcome(user_name)
```
