```python
#Even Odd Counter
count = int(input("How many numbers do you want to enter? "))

even = 0
odd = 0

for i in range(count):
    num = int(input("Enter a number: "))
    
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even numbers:", even)
print("Odd numbers:", odd)
```

```python
# with showing even and odd numbers 

count = int(input("How many numbers do you want to enter? "))

even_numbers = []  # list to store even numbers
odd_numbers = []   # list to store odd numbers

for i in range(count):
    num = int(input("Enter a number: "))
    
    if num % 2 == 0:
        even_numbers.append(num)  # add number to even list
    else:
        odd_numbers.append(num)   # add number to odd list

# Print count and numbers
print("Even numbers:", len(even_numbers), "->", even_numbers)
print("Odd numbers:", len(odd_numbers), "->", odd_numbers)
```

```python
# Even-Odd Counter with Numbers
# This program counts how many even and odd numbers a user enters
# and also shows the actual numbers themselves.

# Step 1: Ask user how many numbers they want to enter
count = int(input("How many numbers do you want to enter? "))

# Step 2: Create empty lists to store even and odd numbers
even_numbers = []  # stores all even numbers
odd_numbers = []   # stores all odd numbers

# Step 3: Loop through all numbers entered by the user
for i in range(count):
    num = int(input("Enter a number: "))  # take input from user
    
    # Step 3a: Check if the number is even
    if num % 2 == 0:
        even_numbers.append(num)  # add number to even list
    else:
        odd_numbers.append(num)   # otherwise, add number to odd list

# Step 4: Print results
# len(even_numbers) gives count of even numbers
# even_numbers shows the actual even numbers
print("Even numbers:", len(even_numbers), "->", even_numbers)
print("Odd numbers:", len(odd_numbers), "->", odd_numbers)

# Example output:
# Enter numbers: 1, 2, 8
# Even numbers: 2 -> [2, 8]
# Odd numbers: 1 -> [1]
```
