```python
# calculator
num1 = int(input("Enter your first number:"))
num2 = int(input("Enter your second number:"))
choice = input("Choose operation (+, -, *, /):")
if choice == "+":
    result = num1 + num2
elif choice == "-":
    result = num1 - num2
elif choice == "*":
    result = num1 * num2
elif choice == "/":
    result = num1 / num2
else:
    print("Invalid choice!")
    
```
